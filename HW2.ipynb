```python
'''1. Як вхідні дані запитайте ціле число. Якщо воно ділиться на 3,
виведіть "foo"; якщо воно ділиться на 5, виведіть "bar"; якщо воно ділиться
на обидва, виведіть "ham" (а не "foo" або "bar").'''
number = int(input('Enter integer: ').strip())
if (number % 3 == 0) and (number % 5 == 0):
    print('ham') 
elif number % 5 == 0:
    print('bar')
elif number % 3 == 0:
    print('foo')
```

```python
# 2. Як вхідні дані запитайте два числа та виведіть яке з них менше і яке більше
number1 = int(input('Enter first number: ').strip())
number2 = int(input('Enter second number: ').strip())
if number1 > number2:
    print(f'{number1} is bigger then {number2}')
elif number1 < number2:
    print(f'{number1} is less then {number2}') 
```

```python
# 3. Як вхідні дані запитайте три числа і виведіть найменше, середнє та найбільше. Припустимо, всі вони різні
number1 = int(input('Enter first number: ').strip())
number2 = int(input('Enter second number: ').strip())
number3 = int(input('Enter third number: ').strip())

numbers = [number1, number2, number3]
numbers.sort()

min_value = numbers[0]
middle_value = numbers[1]
max_value = numbers[2]

print('Minimum value is - ' + str(min_value))
print('Middle value is - ' + str(middle_value))
print('Maximum value is - ' + str(max_value))
```

```python
'''4. Зіграйте у гру Fizz-Buzz: виведіть усі числа від 1 до 100; якщо число
ділиться на 3, замість числа виведіть "fizz". Якщо воно ділиться на 5,
замість числа виведіть "Buzz". Якщо воно ділиться на обидва, виведіть
"fizz buzz" замість числа.'''
for i in range(1, 101):
    if (i % 3 == 0) and (i % 5 == 0):
        print('fizz buzz')
    elif i % 5 == 0:
        print('Buzz')
    elif i % 3 == 0:
        print('fizz')
    else:
        print(i)
```

```python
'''5. Зіграйте у гру 7-boom: виведіть усі числа від 1 до 100; якщо число ділиться
на 7 або містить цифру 7, виведіть "BOOM" замість числа'''
for number in range(1, 101):
    if ('7' in str(number)) or (number % 7 == 0):
        print('BOOM')
    else:
        print(number)
```
